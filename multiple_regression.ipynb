### Multiple Regression

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
dataset= pd.read_excel("E:/PC Backup/Others/fiverr/20200518_multiple_regression/MIS171-2020-T1-A3.xlsx", sheet_name= 'Project1')
dataset.columns
dataset.dtypes

dataset['City']= dataset['City'].astype('category').cat.set_categories(["Sydney", "Melbourne", "Canberra"])

target= "Spend per Transaction"
x= dataset.drop(target, axis=1)
y= dataset[target]

x= pd.get_dummies(x, drop_first= True)
x.columns
x.head(5)

#x, y = np.array(x), np.array(y)
x = sm.add_constant(x)
print(x)
print(y)
```

```python
original_columns= ['const', 'Customer No', 'Age', 'Time spent', 'Health Perception',
       'Food Quality', 'City_Melbourne', 'City_Canberra', 'Gender_Male',
       'Regular Customer_Yes']
drop_columns=[]
columns= (list(set(original_columns) - set(drop_columns)))

model = sm.OLS(y, x[columns])
results = model.fit()
print(results.summary())
print('coefficient of determination:', results.rsquared)
print('adjusted coefficient of determination:', results.rsquared_adj)
print('regression coefficients:', results.params)
```
